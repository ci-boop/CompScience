```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data: hours studied vs. scores obtained
study_hours = np.array([1, 2, 3, 4, 5])
test_scores = np.array([1, 2, 1.3, 3.75, 2.25])

# Function to calculate slope and intercept
def linear_regression(hours, scores):
    n = len(hours)
    avg_hours = np.mean(hours)
    avg_scores = np.mean(scores)

    # Calculate slope (m) and intercept (b)
    numerator = np.sum((hours - avg_hours) * (scores - avg_scores))
    denominator = np.sum((hours - avg_hours) ** 2)
    slope = numerator / denominator
    intercept = avg_scores - slope * avg_hours

    return slope, intercept

# Perform linear regression
slope, intercept = linear_regression(study_hours, test_scores)

# Display results
print(f"Slope (m): {slope:.2f}")
print(f"Intercept (b): {intercept:.2f}")

# Make predictions
predicted_scores = slope * study_hours + intercept

# Plotting the results
plt.scatter(study_hours, test_scores, color='blue', label='Actual Scores')
plt.plot(study_hours, predicted_scores, color='red', label='Best Fit Line')
plt.xlabel('Hours Studied')
plt.ylabel('Scores Obtained')
plt.title('Linear Regression: Study Hours vs. Test Scores')
plt.legend()
plt.grid(True)
plt.show()
```
